```python
import pandas as pd

# Load the Excel file
file_path = r"C:\Users\sales6\Desktop\Work\C019 item code\8915 CB.xlsx"
df = pd.read_excel(file_path, sheet_name ="C019")

# Ensure the target column exists
col_name = "客戶產品代號(P/N)"
if col_name in df.columns:
    # Extract text before the second "."
    df["unique_item_code"] = df[col_name].astype(str).apply(lambda x: ".".join(x.split(".")[:2]) )
     # Get unique extracted values
    unique_df = df.drop_duplicates(subset=["unique_item_code"])
    unique_df.to_excel("8915 CB unique.xlsx")
    # Print result
    print(unique_df)

    
```
